# 1. Basic SVM
---

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets, metrics
```

```python
iris = datasets.load_iris()
iris.data.shape
```

```python
X = iris.data[:,:2]
Y = iris.target

print(X.shape)
print(Y.shape)
```

```python
C = 0.22
model = svm.SVC(kernel = 'linear', C = C)
model.fit(X,Y)
```

```python
pred = model.predict(X)
pred
```

```python
print(Y,'\n')
print(pred)
```

```python
metrics.confusion_matrix(Y,pred)
```

```python
accuracy = metrics.accuracy_score(Y,pred)
print("정확도:", accuracy)

precision = metrics.precision_score(Y,pred,average='macro')
print("정밀도:", precision)

recall = metrics.recall_score(Y,pred,average='macro')
print("재현율:", recall)

f1 = metrics.f1_score(Y,pred,average='macro')
print("f1 점수:", f1)
```

#### plot_contours (등치선)

- plotting하고자 원하는 영역에 대한 mesh_grid 생성 
  - 범위를 설정하고 numpy 내의 함수(ex, linespace)를 통해 균일한 크기로 쪼갬
  - 해당 영역을 이용해 mesh를 생성 np.meshgrid(x,y)

  

```python
def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
# 모델 정의&피팅
C = 1.0 #regularization parameter
models = (svm.SVC(kernel='linear', C=C),
          svm.LinearSVC(C=C, max_iter=10000),
          svm.SVC(kernel='rbf', gamma=0.7, C=C),
          svm.SVC(kernel='poly', degree=3, gamma='auto', C=C))
models = (clf.fit(X, Y) for clf in models)
```

```python
# plot title 형성
titles = ('SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel')
```

```python
# plot 그리기

fig, sub = plt.subplots(2, 2, figsize=(12,12))
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy,
                  cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X0, X1, c=Y, cmap=plt.cm.coolwarm, s=40, edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('Sepal length')
    ax.set_ylabel('Sepal width')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```
